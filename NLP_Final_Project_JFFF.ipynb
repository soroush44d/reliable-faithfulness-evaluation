```python
import re

def extract_scores(text):
    q1_pattern = r"'Q1':\s(\d+\.\d+)"
    q2_pattern = r"'Q2':\s(\d+\.\d+)"

    q1_matches = re.findall(q1_pattern, text)
    q2_matches = re.findall(q2_pattern, text)

    q1_scores = [float(score) for score in q1_matches] if q1_matches else []
    q2_scores = [float(score) for score in q2_matches] if q2_matches else []

    return q1_scores, q2_scores


def genai_query(original, summary, api_key):
  model_name = 'gemini-1.5-flash-8b'
  genai.configure(api_key=api_key)
  model = genai.GenerativeModel(model_name)
  query = f"""
  Based on the original text and the summary, provide scores between 0 and 1 for the following questions in this format: ('Q1': result, 'Q2': result) .
  no need to other Explanation and Each question's score should reflect its individual likelihood, independent of the other question's score.
  Q1 = Is all information in the summary accurate based on the source text?
  Q2 = Does the summary contain information that contradicts the source text?
  You Have 4 pair Original Text and Summary to evaluate so return 4 scores in this format.
  """

  for i in range(len(original)):
    query += f"""
    -------------------------
    pair {i+1})
    Original Text: {original[i]}
    Summary: {summary[i]}
    """
  response = model.generate_content(query)

  print(response.text)
  return extract_scores(response.text)
```

```python
model_name = "distilgpt2" # Or We Can use gpt 2
Entropy_Tokenizer = AutoTokenizer.from_pretrained(model_name)
Entropy_model = AutoModelForCausalLM.from_pretrained(model_name)

def Token_Level_Entropy(text, Entropy_model, Entropy_Tokenizer):
  inputs = Entropy_Tokenizer(text, return_tensors="pt")
  with torch.no_grad():
      # Pass the input tensors directly to the model
      outputs = Entropy_model(**inputs)
      logits = outputs.logits

  shift_logits = logits[:, :-1, :].contiguous()

  log_probs = torch.log_softmax(shift_logits, dim=-1)
  # Avoid log(0) by adding a small epsilon
  epsilon = 1e-10
  token_entropies_tensor = -torch.sum(torch.exp(log_probs) * torch.log(torch.exp(log_probs) + epsilon), dim=-1)


  token_entropies = token_entropies_tensor.flatten().tolist()
  average_entropy = np.mean(token_entropies)

  return float(average_entropy)
```

```python
from tqdm import tqdm
import pandas as pd

def find_Q(df_train, api_key_list):
    if not isinstance(df_train, pd.DataFrame) or df_train.empty:
        raise ValueError("Invalid DataFrame: Must be a non-empty pandas DataFrame")
    if 'document' not in df_train.columns or 'summary' not in df_train.columns:
        raise ValueError("DataFrame must contain 'document' and 'summary' columns")
    if not api_key_list:
        raise ValueError("API key list cannot be empty")

    original_list = list(df_train['document'])
    summary_list = list(df_train['summary'])
    batch_size = 4
    steps_per_batch = len(original_list) // batch_size
    rem = len(original_list) % batch_size
    key_count = len(api_key_list)
    all_q1_scores = []
    all_q2_scores = []

    print("Steps:", steps_per_batch)
    print("Remainder:", rem)

    df_result = df_train.copy()

    total_steps = steps_per_batch + (1 if rem > 0 else 0)

    with open("ans_values.txt", "w", encoding="utf-8") as f:
        for i in tqdm(range(45, steps_per_batch), desc="Processing batches", total=total_steps):
            api_key_index = i % key_count
            origin_batch = original_list[batch_size * i : batch_size * (i + 1)]
            summary_batch = summary_list[batch_size * i : batch_size * (i + 1)]
            try:
                ans = genai_query(origin_batch, summary_batch, api_key_list[api_key_index])
                q1_scores, q2_scores = ans
                all_q1_scores.extend(q1_scores)
                all_q2_scores.extend(q2_scores)
                f.write(f"Step {i + 1}:\n")
                f.write(f"{ans}\n\n")
                f.write("*" * 100 + "\n")
            except Exception as e:
                print(f"Error in batch {i + 1}: {e}")
                raise

        if rem > 0:
            api_key_index = steps_per_batch % key_count
            origin_batch = original_list[-rem:]
            summary_batch = summary_list[-rem:]
            try:
                ans = genai_query(origin_batch, summary_batch, api_key_list[api_key_index])
                q1_scores, q2_scores = ans
                all_q1_scores.extend(q1_scores)
                all_q2_scores.extend(q2_scores)
                f.write(f"Remaining Step:\n")
                f.write(f"{ans}\n\n")
                f.write("*" * 100 + "\n")
            except Exception as e:
                print(f"Error in remaining batch: {e}")
                raise

    if len(all_q1_scores) != len(df_result) or len(all_q2_scores) != len(df_result):
        raise ValueError("Length of Q1/Q2 scores does not match DataFrame rows")

    df_result["Q1"] = all_q1_scores
    df_result["Q2"] = all_q2_scores

    return df_result

try:
    df_qq = find_Q(df_aggregated, api_key_list)
except Exception as e:
    print(f"Error: {e}")
```

```python
# Approch B
import spacy
from typing import List, Tuple

!python -m spacy download en_core_web_md
!python -m spacy download en_core_web_lg

large_model = "en_core_web_lg"
medium_model = "en_core_web_md"
nlp = spacy.load(large_model)
def extract_entities(text: str, nlp) -> List[Tuple[str, str]]:

    doc = nlp(text)

    entities = []
    seen_entities = set()
    for ent in doc.ents:
        entity_text = ent.text.strip()
        entity_type = ent.label_
        if entity_text and (entity_text, entity_type) not in seen_entities:
            entities.append((entity_text, entity_type))
            seen_entities.add((entity_text, entity_type))

    return entities
```

```python
def get_full_phrase(token: spacy.tokens.Token, cache: dict = None) -> str:
    if not token:
        return ""
    if cache is None:
        cache = {}
    if token in cache:
        return cache[token]
    subtree_tokens = [t for t in token.subtree if not t.is_punct]
    phrase = " ".join(t.text for t in subtree_tokens)
    cache[token] = phrase.strip()
    return phrase.strip()

def extract_svo(text: str, nlp) -> List[Tuple[str, str, str]]:
    raw_entities = extract_entities(text, nlp)
    entity_types = {entity.lower(): ent_type for entity, ent_type in raw_entities}

    doc = nlp(text)
    svo_triples = []
    phrase_cache = {}

    for sent in doc.sents:
        resolved_tokens = {}
        last_noun_or_propn_token = None

        for token in sent:
            if token.pos_ in ("NOUN", "PROPN"):
                last_noun_or_propn_token = token
                resolved_tokens[token] = get_full_phrase(token, phrase_cache)
            elif token.pos_ == "PRON" and last_noun_or_propn_token:
                if token.text.lower() in ["he", "she", "it", "they", "him", "her", "them"]:
                    resolved_tokens[token] = get_full_phrase(last_noun_or_propn_token, phrase_cache)
                else:
                    resolved_tokens[token] = token.text
            else:
                resolved_tokens[token] = token.text

        verbs = [token for token in sent if token.pos_ == "VERB" or
                 (token.pos_ == "AUX" and token.dep_ == "ROOT" and list(token.children))]

        for verb in verbs:
            subject_token = next((child for child in verb.children if child.dep_ in ("nsubj", "nsubjpass")), None)
            object_token = None

            if subject_token and subject_token.dep_ == "nsubjpass":
                agent = next((child for child in verb.children if child.dep_ == "agent"), None)
                if agent:
                    object_token = next((grand_child for grand_child in agent.children if grand_child.dep_ == "pobj"), None)

            if not object_token:
                object_token = next((child for child in verb.children if child.dep_ in ("dobj", "pobj")), None)

            subject_text = resolved_tokens.get(subject_token, get_full_phrase(subject_token, phrase_cache)) if subject_token else None
            object_text = resolved_tokens.get(object_token, get_full_phrase(object_token, phrase_cache)) if object_token else None

            if object_token and object_text == object_token.text:
                object_text = get_full_phrase(object_token, phrase_cache)

            if object_text and object_text.lower() in entity_types and entity_types[object_text.lower()] in ("GPE", "LOC"):
                if object_token and object_token.dep_ == "pobj":
                    object_text = None

            if subject_text and verb.text and object_text:
                svo_triples.append((subject_text.strip(), verb.text.strip(), object_text.strip()))

    return svo_triples
```

```python
from sentence_transformers import SentenceTransformer, util
from typing import Optional,Dict

sim_model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
def compute_semantic_similarity(word1: str, word2: str, model=sim_model) -> Optional[float]:

    if not word1 or not word2 or not isinstance(word1, str) or not isinstance(word2, str):
        return None

    embedding1 = model.encode(word1, convert_to_tensor=True)
    embedding2 = model.encode(word2, convert_to_tensor=True)

    similarity = util.cos_sim(embedding1, embedding2).item()

    return max(0.0, min(1.0, similarity))


def compare_entities(source_text: str, summary_text: str, nlp):
    source_entities = extract_entities(source_text, nlp)
    summary_entities = extract_entities(summary_text, nlp)

    supported_count = 0

    with open("entity_comparison_log.txt", "a") as f:
      f.write(f"--- Comparing Summary: '{summary_text}' with Source: '{source_text}' ---\n")
      for entity, entity_type in summary_entities:
          result = {
              "entity": entity,
              "type": entity_type,
              "supported": False,
              "match_type": "none",
              "reason": "",
              "cosine_similarity": None,
              "matched_to": None
          }

          # Check for exact match
          for src_entity, src_type in source_entities:
              if entity.lower() == src_entity.lower():
                  result["supported"] = True
                  result["match_type"] = "exact"
                  result["matched_to"] = src_entity
                  supported_count += 1
                  break

          # If no exact match, check for semantic match
          if not result["supported"]:
              max_similarity = 0.0
              best_match = None
              for src_entity, src_type in source_entities:
                  if entity_type == src_type:
                      similarity = compute_semantic_similarity(entity, src_entity)
                      if similarity is not None and similarity > max_similarity:
                          max_similarity = similarity
                          best_match = src_entity

              if max_similarity is not None and max_similarity > 0.8:
                  result["supported"] = True
                  result["match_type"] = "semantic"
                  result["cosine_similarity"] = max_similarity
                  result["matched_to"] = best_match
                  supported_count += 1
              else:
                  result["reason"] = "Unsupported - no exact or semantic match (cosine similarity <= 0.8)"
                  result["cosine_similarity"] = max_similarity if max_similarity is not None else None
                  result["matched_to"] = best_match if best_match else None
                  f.write(f"  Entity: {result['entity']} (Type: {result['type']}) - Supported: {result['supported']} - Match Type: {result['match_type']} - Matched To: {result['matched_to']} - Cosine Similarity: {result['cosine_similarity']} - Reason: {result['reason']}\n")

          # Write every entity comparison result to the log file
          if result["supported"]:
              f.write(f"  Entity: {result['entity']} (Type: {result['type']}) - Supported: {result['supported']} - Match Type: {result['match_type']} - Matched To: {result['matched_to']}\n")


      f.write("---\n")


    # Calculate Entity Precision
    entity_precision = supported_count / len(summary_entities) if summary_entities else 0.0

    # Return only the entity precision
    return entity_precision
```

```python
from typing import Dict, List

def compare_svo(source_text: str, summary_text: str, nlp) -> Dict:
    # Extract entities for type checking
    source_entities = extract_entities(source_text, nlp)
    summary_entities = extract_entities(summary_text, nlp)
    entity_types = {entity: ent_type for entity, ent_type in source_entities + summary_entities}

    # Extract S-V-O triples
    source_triples = extract_svo(source_text, nlp)
    summary_triples = extract_svo(summary_text, nlp)

    supported_count = 0

    with open("svo_comparison_log.txt", "a") as f:
        f.write(f"--- Comparing Summary: '{summary_text}' with Source: '{source_text}' ---\n")
        f.write(f"Source triples: {source_triples}\n")
        f.write(f"Summary triples: {summary_triples}\n")

        # Compare each summary triple with source triples
        for s, v, o in summary_triples:
            result = {
                "triple": f"{s}-{v}-{o}",
                "supported": False,
                "match_type": "none",
                "reason": "",
                "semantic_details": {},
                "matched_to": None
            }

            # Check for exact match
            for src_s, src_v, src_o in source_triples:
                if (s.lower(), v.lower(), o.lower()) == (src_s.lower(), src_v.lower(), src_o.lower()):
                    result["supported"] = True
                    result["match_type"] = "exact"
                    result["matched_to"] = f"{src_s}-{src_v}-{src_o}"
                    supported_count += 1
                    break

            # If no exact match, check for semantic match or hallucinations
            if not result["supported"]:
                max_similarity = 0.0
                best_match_triple = None
                semantic_details = {}

                # Initialize similarities list
                similarities = []  # Always initialize here to avoid UnboundLocalError

                for src_s, src_v, src_o in source_triples:
                    # Compare subject (only if same entity type or not an entity)
                    if s.lower() != src_s.lower() and (s not in entity_types or src_s not in entity_types or entity_types.get(s) == entity_types.get(src_s)):
                        sim_s = compute_semantic_similarity(s, src_s)
                        if sim_s is not None:
                            similarities.append(("subject", s, src_s, sim_s))
                    # Compare verb
                    if v.lower() != src_v.lower():
                        sim_v = compute_semantic_similarity(v, src_v)
                        if sim_v is not None:
                            similarities.append(("verb", v, src_v, sim_v))
                    # Compare object (only if same entity type or not an entity)
                    if o.lower() != src_o.lower() and (o not in entity_types or src_o not in entity_types or entity_types.get(o) == entity_types.get(src_o)):
                        sim_o = compute_semantic_similarity(o, src_o)
                        if sim_o is not None:
                            similarities.append(("object", o, src_o, sim_o))

                    # Check if differing components have similarity > 0.8
                    if similarities:
                        all_similar = all(sim > 0.8 for _, _, _, sim in similarities)
                        if all_similar and len(similarities) <= 2:  # Allow up to 2 components to differ
                            current_similarity = sum(sim for _, _, _, sim in similarities) / len(similarities)
                            if current_similarity > max_similarity:
                                max_similarity = current_similarity
                                best_match_triple = (src_s, src_v, src_o)
                                semantic_details = {comp: {f"{w1}": f"{w2}", "cosine_similarity": sim} for comp, w1, w2, sim in similarities}

                # Check if semantic match
                if max_similarity > 0.8:
                    result["supported"] = True
                    result["match_type"] = "semantic"
                    result["semantic_details"] = semantic_details
                    result["matched_to"] = f"{best_match_triple[0]}-{best_match_triple[1]}-{best_match_triple[2]}" if best_match_triple else None
                    supported_count += 1
                else:
                    # Check for intrinsic or extrinsic hallucination
                    intrinsic = False
                    extrinsic = False
                    if similarities:  # Now safe to check
                        for comp, w1, w2, sim in similarities:
                            if (comp in ("subject", "object") and w1 in entity_types and w2 in entity_types and entity_types[w1] == entity_types[w2]):
                                if 0.45 <= sim <= 0.8:
                                    intrinsic = True
                                    break
                                elif sim < 0.45:
                                    extrinsic = True
                    result["reason"] = ("Intrinsic Hallucination" if intrinsic else
                                       "Extrinsic Hallucination" if extrinsic else
                                       "Unsupported - no exact or semantic match (cosine similarity <= 0.8)")
                    result["semantic_details"] = semantic_details if semantic_details else {}
                    result["matched_to"] = f"{best_match_triple[0]}-{best_match_triple[1]}-{best_match_triple[2]}" if best_match_triple else None

            # Write every triple comparison result to the log file
            f.write(f"  Triple: {result['triple']} - Supported: {result['supported']} - Match Type: {result['match_type']} - Matched To: {result['matched_to']} - Semantic Details: {result['semantic_details']} - Reason: {result['reason']}\n")

        f.write("---\n")

    # Calculate Triple Precision
    triple_precision = supported_count / len(summary_triples) if summary_triples else 0.0

    # Return only the triple precision
    return triple_precision
```

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer
from sentence_transformers import SentenceTransformer
import torch
import numpy as np
```

```python
nli_model_name = "cross-encoder/nli-deberta-v3-base"
sbert_model_name = "all-MiniLM-L6-v2"

nli_tokenizer = AutoTokenizer.from_pretrained(nli_model_name)
nli_model = AutoModelForSequenceClassification.from_pretrained(nli_model_name)

sbert_model = SentenceTransformer(sbert_model_name)

def calculate_nli_score(original_text, summary_text):
    inputs = nli_tokenizer(f"{original_text} [SEP] {summary_text}",
                           return_tensors="pt",
                           padding=True,
                           truncation=True,
                           max_length=512)

    with torch.no_grad():
        outputs = nli_model(**inputs)
        logits = outputs.logits

    probs = torch.softmax(logits, dim=1)
    p_entail = probs[0][2].item()
    p_contradict = probs[0][0].item()

    nli_score = p_entail - p_contradict
    return nli_score
```

```python
def calculate_topic_drift(original_text, summary_text):
    embeddings = sbert_model.encode([original_text, summary_text])
    v_doc = embeddings[0] / np.linalg.norm(embeddings[0])
    v_sum = embeddings[1] / np.linalg.norm(embeddings[1])

    cosine_sim = np.dot(v_doc, v_sum)

    drift = 1 - cosine_sim
    return drift
```

```python
def Score_C(original_text, summary_text):
    nli_score = calculate_nli_score(original_text, summary_text)

    drift = calculate_topic_drift(original_text, summary_text)

    return nli_score, drift
```

```python
from tqdm.auto import tqdm

def find_nli_dirft(df_train):
    nli_scores = []
    drift_scores = []

    for index, row in tqdm(df_train.iterrows(), total=len(df_train), desc="Calculating NLI and Drift Scores"):
        original_text = row['document']
        summary_text = row['summary']
        nli_score, drift_score = Score_C(original_text, summary_text)
        nli_scores.append(nli_score)
        drift_scores.append(drift_score)

    df_train['nli_score'] = nli_scores
    df_train['drift_score'] = drift_scores

    return df_train

# Assuming df_aggregated is the DataFrame you want to process
df_aggregated = find_nli_dirft(df_aggregated)
```

**Calibration: Temperature Scaling**


Train a Base Probability Model: For each raw score (e.g., Q1, NLI Score), a simple logistic regression model is trained. This model learns to predict the Faithful ground truth based on that single raw score. It essentially converts the raw score into a preliminary probability.

Optimize Temperature (T):

> 150 sample

The temperature parameter is a positive scalar (T > 0) that divides the raw logits (log-odds) of the base model's predictions.

>If T = 1, no scaling occurs.

>If T > 1, the probabilities become "softer" (less confident).

>If T < 1, the probabilities become "sharper" (more confident).

>The optimal T is found by minimizing the Negative Log-Likelihood (NLL) on a separate validation set. This ensures that the calibrated probabilities are as close as possible to the true probabilities.

>Apply Optimal Temperature: Once the optimal T is found for each score, it's applied to the entire dataset's raw scores (or their corresponding logits) to produce calibrated probabilities between 0 and 1.

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from scipy.optimize import minimize

df = original_df.copy()

score_columns = [
    'entity_score', 'nli_score', 'drift_score',
    'Q1', 'Q2', 'token_entropy'
]

target_columns = ['Hallucination']

def temperature_scaling(scores, model, temperature=1.5):
    probs = model.predict_proba(scores.reshape(-1, 1))[:, 1]
    logits = np.log(probs / (1 - probs))
    calibrated_logits = logits / temperature
    calibrated_probs = 1 / (1 + np.exp(-calibrated_logits))
    return np.clip(calibrated_probs, 0, 1)

def nll(temperature, X_val, y_val, model):
    calibrated_probs = temperature_scaling(X_val, model, temperature)
    calibrated_probs = np.clip(calibrated_probs, 1e-10, 1 - 1e-10)
    return -np.mean(y_val * np.log(calibrated_probs) + (1 - y_val) * np.log(1 - calibrated_probs))

for score_col in score_columns:
    for target_col in target_columns:
        df_sub = df[[score_col, target_col]].dropna()
        if len(df_sub) < 150:
            print(f"Skipping {score_col} → {target_col}: Insufficient data (<150)")
            continue

        # Use only the first 150 samples for calibration
        df_cal = df_sub.iloc[:150].copy()
        df_train = df_cal.iloc[:120].copy()  # 120 samples for training the model
        df_val = df_cal.iloc[120:].copy()    # 30 samples for validation

        X_train = df_train[score_col].values.reshape(-1, 1)
        y_train = df_train[target_col].values
        X_val = df_val[score_col].values.reshape(-1, 1)
        y_val = df_val[target_col].values

        model = LogisticRegression()
        y_binary_train = (y_train > np.median(y_train)).astype(int)
        model.fit(X_train, y_binary_train)

        result = minimize(nll, x0=1.0, args=(X_val, (y_val > np.median(y_val)).astype(int), model), bounds=[(0.1, 10)])
        optimal_t = result.x[0]
        print(f"Optimal T for {score_col} → {target_col}: {optimal_t:.4f}")

        try:
            X_all = df[score_col].values.reshape(-1, 1)
            calibrated_values = temperature_scaling(X_all, model, temperature=optimal_t)
            df[f"{score_col}_cal"] = calibrated_values
        except Exception as e:
            print(f"Error in calibrating {score_col} → {target_col}: {e}")

# df.to_csv("dataset_with_calibrated_scores.csv", index=False)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
import itertools
from tqdm import tqdm

# List of calibrated features that will be combined
features = [
    'Q1_cal',
    'Q2_cal',
    'token_entropy_cal',
    'entity_score_cal',
    'nli_score_cal',
    'drift_score_cal'
]

# Ground Truth targets - Changed to include ONLY 'Hallucination'
targets = ['Hallucination']

# Define the search space for weights
# Negative weights only for Q2_cal, token_entropy_cal, drift_score_cal
negative_features = ['Q1_cal', 'token_entropy_cal', 'nli_score_cal']
positive_features = ['Q2_cal', 'entity_score_cal' , 'drift_score_cal']

weight_steps_negative = np.arange(-1.0, 1.01, 0.20)  # -1.0 to 1.0
weight_steps_positive = np.arange(0.0, 1.01, 0.20)   # 0.0 to 1.0

# Generate all possible combinations of weights
all_weight_combinations = list(itertools.product(
    *(weight_steps_positive if f in positive_features else weight_steps_negative for f in features)
))

best_correlation = -1.0
best_weights = None

# Perform Grid Search on df_grid with tqdm
try:
    _ = df_grid # Check if df_grid exists
except NameError:
    print("Error: df_grid is not defined. Please ensure your grid search data is loaded into df_grid.")
    # You might want to exit or handle this error more gracefully depending on your script's flow
    exit() # Exits the script if df_grid is not found


for weights_tuple in tqdm(all_weight_combinations, desc="Grid Search Progress"):
    current_weights = dict(zip(features, weights_tuple))

    df_grid['S_fused'] = 0.0
    for feature in features:
        df_grid['S_fused'] += df_grid[feature] * current_weights[feature]

    # Calculate correlation for the single target
    correlation, _ = pearsonr(df_grid['S_fused'], df_grid['Hallucination'])

    if correlation > best_correlation:
        best_correlation = correlation
        best_weights = current_weights

print("Best Weights found:")
for feature, weight in best_weights.items():
    print(f"  {feature}: {weight:.4f}")
print(f"\nMaximum Pearson Correlation with 'Hallucination' score: {best_correlation:.4f}")

# Apply weights to test set
try:
    _ = df_final_test # Check if df_final_test exists
except NameError:
    print("\nError: df_final_test is not defined. Please ensure your test data is loaded into df_final_test.")
    # You might want to exit or handle this error more gracefully
    exit() # Exits the script if df_final_test is not found


df_final_test['S_fused'] = 0.0
for feature in features:
    df_final_test['S_fused'] += df_final_test[feature] * best_weights[feature]
test_correlations = {}
for target_col in targets:
    test_correlations[target_col] = pearsonr(df_final_test['S_fused'], df_final_test[target_col])[0]
print(f"\nTest Set Correlations:")
for target_col, corr in test_correlations.items():
    print(f"  {target_col} Correlation: {corr:.4f}")
```
